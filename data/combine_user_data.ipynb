```python
import pandas as pd

# Tạo dataframe master_data rỗng để lưu kết quả
master_data = pd.DataFrame()

def merge_style_1(column_master, column_current, same_order_id=False):
    if same_order_id:
        arr = column_current.split()
        arr = arr[1:]  # Cập nhật lại mảng sau khi loại bỏ phần tử đầu tiên
        return str(column_master) + ' ' + str(' '.join(arr))
    else:
        return str(column_master) + ' ' + str(column_current)

def merge_style_2(column_master, column_current, same_order_id=False):
    if same_order_id:
        return str(column_master) + '_' + str(column_current)
    else:
        return str(column_master) + ' ' + str(column_current)
```

```python
master_data = pd.DataFrame()  # Khởi tạo master_data bên ngoài vòng lặp để giữ lại dữ liệu giữa các file

# Vòng lặp qua từng file
for i in range(0, 33500001, 500000):
    # Đọc file hiện tại
    file_path = f'/content/data/user_data_{i}.csv'
    current_data = pd.read_csv(file_path)

    # Kiểm tra nếu master_data đã có dữ liệu
    if not master_data.empty:
        # Kiểm tra user_id cuối của master_data và user_id đầu của current_data
        last_user_id_master = master_data.iloc[-1].user_id
        first_user_id_current = current_data.iloc[0].user_id

        if last_user_id_master == first_user_id_current:
            master_order_ids_list = master_data.iloc[-1].order_ids
            current_order_ids_list = current_data.iloc[0].order_ids
            last_master_order_id = master_order_ids_list.split()[-1]
            current_first_order_id = current_order_ids_list.split()[0]
            flag = (last_master_order_id == current_first_order_id)

            # Merge các cột tương ứng
            merged_order_ids = merge_style_1(master_data.iloc[-1].order_ids, current_data.iloc[0].order_ids, same_order_id=flag)
            merged_order_numbers = merge_style_1(master_data.iloc[-1].order_numbers, current_data.iloc[0].order_numbers, same_order_id=flag)
            merged_order_dows = merge_style_1(master_data.iloc[-1].order_dows, current_data.iloc[0].order_dows, same_order_id=flag)
            merged_order_hours = merge_style_1(master_data.iloc[-1].order_hours, current_data.iloc[0].order_hours, same_order_id=flag)
            merged_days_since_prior_orders = merge_style_1(master_data.iloc[-1].days_since_prior_orders, current_data.iloc[0].days_since_prior_orders, same_order_id=flag)
            merged_product_ids = merge_style_2(master_data.iloc[-1].product_ids, current_data.iloc[0].product_ids, same_order_id=flag)
            merged_aisle_ids = merge_style_2(master_data.iloc[-1].aisle_ids, current_data.iloc[0].aisle_ids, same_order_id=flag)
            merged_department_ids = merge_style_2(master_data.iloc[-1].department_ids, current_data.iloc[0].department_ids, same_order_id=flag)
            merged_reorders = merge_style_2(master_data.iloc[-1].reorders, current_data.iloc[0].reorders, same_order_id=flag)
            merged_eval_set = current_data.iloc[0].eval_set

            # Cập nhật dữ liệu đã merge vào master_data
            master_data.at[master_data.index[-1], 'order_ids'] = merged_order_ids
            master_data.at[master_data.index[-1], 'order_numbers'] = merged_order_numbers
            master_data.at[master_data.index[-1], 'order_dows'] = merged_order_dows
            master_data.at[master_data.index[-1], 'order_hours'] = merged_order_hours
            master_data.at[master_data.index[-1], 'days_since_prior_orders'] = merged_days_since_prior_orders
            master_data.at[master_data.index[-1], 'product_ids'] = merged_product_ids
            master_data.at[master_data.index[-1], 'aisle_ids'] = merged_aisle_ids
            master_data.at[master_data.index[-1], 'department_ids'] = merged_department_ids
            master_data.at[master_data.index[-1], 'reorders'] = merged_reorders
            master_data.at[master_data.index[-1], 'eval_set'] = merged_eval_set

            # Loại bỏ dòng đầu tiên của current_data (đã được merge vào master_data)
            current_data = current_data.iloc[1:]
            # Nối current_data vào master_data
            master_data = pd.concat([master_data, current_data], ignore_index=True)
        else:
            # Nếu user_id không trùng, nối current_data vào master_data
            master_data = pd.concat([master_data, current_data], ignore_index=True)

    else:
        # Nếu master_data chưa có dữ liệu, gán dữ liệu đầu tiên từ current_data vào master_data
        master_data = current_data

```

```python
# Lưu kết quả cuối cùng vào file
master_data.to_csv('merged_user_data.csv', index=False)
```
